## Задание 2:  ДТП с жертвами (folium)

Показать участки дорог, где количество смертельных случаев больше 10 человек. 

Добавить количество жертв (число погибших) на карту в виде подписей к маркерам, можно использовать folium.DivIcon для создания текстовых меток. 
Добавьте тултипы - всплывающие сообщения (при наведении), на которых отображается причина происшествия, дата аварии в формате (dd.mm.YYYY), тип происшествия. 

Метод: для folium метод CircleMarker. 


```python
import folium

# Создаем карту
m = folium.Map(location=[55.751244, 37.618423], zoom_start=5)

# Фильтруем данные (ДТП с погибшими)
fatal_crashes = df[df['fatalities_amount'] > 10].dropna(subset=['latitude', 'longitude'])

# Добавляем круговые маркеры с подписями
for _, row in fatal_crashes.iterrows():
    # Круговой маркер
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row['fatalities_amount'] * 2,  # Размер зависит от числа жертв
        color='red',
        fill=True,
        fill_opacity=0.7,
        weight=1,
        popup=f"Причина происшествия: {row['crash_reason']}<br>Дата: {row['crash_date']}<br> Тип происшествия: {row['crash_type_name']}"
    ).add_to(m)
    
    # Текст с количеством жертв
    folium.map.Marker(
        [row['latitude'], row['longitude']],
        icon=folium.DivIcon(
            icon_size=(30, 30),
            html=f'<div style="font-size: 10pt; color: white; text-align: center; background: transparent;">{int(row["fatalities_amount"])}</div>'
        )
    ).add_to(m)

# Сохраняем карту
m
```

## Задание 3: Причины и типы самых ужастных происшествий
Изменить карту из задания 2, добавить стили для маркеров - отдельные типы происшествий должны быть отображаться отдельным цветов.
В легенде  вывести цвет - тип происшествия. 


```python
import folium

# Создаем карту
m = folium.Map(location=[55.751244, 37.618423], zoom_start=5)

# Цвета для разных типов ДТП
crash_type_colors = {
    'Наезд на пешехода': 'red',
    'Столкновение': 'blue',
    'Опрокидывание': 'green',
    'Наезд на препятствие': 'orange',
    'Наезд на животное': 'purple',
    'Наезд на стоящее ТС':'lightgray',
    'Съезд с дороги': 'cadetblue',
    'Другое': 'gray'
}

# Фильтруем данные (ДТП с более чем 10 погибшими)
fatal_crashes = df[df['fatalities_amount'] > 10].dropna(subset=['latitude', 'longitude'])

# Добавляем круговые маркеры с подписями
for _, row in fatal_crashes.iterrows():
    crash_type = row['crash_type_name']
    color = crash_type_colors.get(crash_type, 'gray')  # По умолчанию серый
    
    # Круговой маркер
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row['fatalities_amount'] * 2,  # Размер зависит от числа жертв
        color=color,
        fill=True,
        fill_opacity=0.7,
        weight=1,
        popup=f"""
        Тип: {crash_type}<br>
        Жертв: {row['fatalities_amount']}<br>
        Дата: {row['crash_date']}<br>
        Причина: {row['crash_reason']}
        """
    ).add_to(m)
    
    # Текст с количеством жертв
    folium.map.Marker(
        [row['latitude'], row['longitude']],
        icon=folium.DivIcon(
            icon_size=(30, 30),
            html=f'<div style="font-size: 10pt; color: white; text-align: center; background: transparent;">{int(row["fatalities_amount"])}</div>'
        )
    ).add_to(m)

# Создаем легенду
legend_html = """
<div style="position: fixed; 
     bottom: 50px; left: 50px; width: 180px; height: auto; 
     border:2px solid grey; z-index:9999; font-size:14px;
     background-color:white;
     padding: 10px;
     ">
     <b>Типы ДТП</b><br>
"""

for crash_type, color in crash_type_colors.items():
    legend_html += f"""
    <span style="color: {color}; font-size: 25px;">■</span> {crash_type}<br>
    """

legend_html += "</div>"

m.get_root().html.add_child(folium.Element(legend_html))

# Сохраняем карту
m
```

## Задание 4: Анализ ДТП по времени суток

Показать динамику аварийности в течение дня.

Выведите три диаграммы:
1. Распределение ДТП по времени суток:
- Столбчатая диаграмма с количеством аварий для каждого периода: ('Ночь (0-6)', 'Утро (6-12)', 'День (12-18)', 'Вечер (18-24)') - разделение на ночь/утро/день/вечер
- Цветовая кодировка периодов суток

2. Типы ДТП по времени суток:
- Накопительная столбчатая диаграмма - какие типы аварий преобладают в разное время

3. Тяжесть ДТП по часам:
- Линейный график средней тяжести аварий
- Отдельные линии для погибших и пострадавших

Вывести легенду по типам ДТП.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.patches import Patch

# Подготовка данных
df['hour'] = pd.to_datetime(df['crash_time']).dt.hour
df['time_of_day'] = pd.cut(df['hour'],
                          bins=[0, 6, 12, 18, 24],
                          labels=['Ночь (0-6)', 'Утро (6-12)', 'День (12-18)', 'Вечер (18-24)'],
                          right=False)

# 1. Распределение ДТП по времени суток
plt.figure(figsize=(12, 6))
time_counts = df['time_of_day'].value_counts().sort_index()
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

plt.subplot(1, 2, 1)
bars = plt.bar(time_counts.index, time_counts.values, color=colors)
plt.title('Общее количество ДТП по времени суток')
plt.xlabel('Время суток')
plt.ylabel('Количество ДТП')
plt.xticks(rotation=45)

# Добавляем значения на столбцы
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

# 2. Распределение типов ДТП по времени суток
plt.subplot(1, 2, 2)
type_counts = df.groupby(['time_of_day', 'crash_type_name']).size().unstack()
type_counts.plot(kind='bar', stacked=True, ax=plt.gca(), 
                color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#8c994b', 
                       '#1c094b', '#00ffff', '#7ffdf4', '#9acd32', '#c0c0c0', '#add8e6','#ffffff',
                      '#000000','#380282','#ffa500','#ffff00', '#123456'])
plt.title('Типы ДТП по времени суток')
plt.xlabel('Время суток')
plt.ylabel('Количество ДТП')
plt.xticks(rotation=45)
plt.legend(title='Тип ДТП', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

# 3. Тяжесть ДТП по часам
plt.figure(figsize=(12, 6))
hourly_severity = df.groupby('hour')[['fatalities_amount', 'victims_amount']].mean()
plt.plot(hourly_severity.index, hourly_severity['fatalities_amount'], 
         label='Среднее число погибших', marker='o', color='#d62728')
plt.plot(hourly_severity.index, hourly_severity['victims_amount'], 
         label='Среднее число пострадавших', marker='o', color='#ff7f0e')
plt.title('Тяжесть ДТП по часам')
plt.xlabel('Час дня')
plt.ylabel('Количество людей')
plt.xticks(range(0, 24))
plt.legend()
plt.grid(True)

plt.show()
```

## Задание 5: Анализ ДТП по сезонам (зима, лето, весна, осень)

Выведите 4 диаграмм:
1. Круговая диаграмма среднего количества ДТП по временам года
2. Столбчатая диаграмма среднего количества погибших по временам года
3. Столбчатая диаграмма среднего количества пострадавших по временам года
4. Линейный график с помесячной статистикой количества ДТП




```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.ticker import MaxNLocator

# Преобразуем дату в datetime и извлекаем месяц
df['crash_date'] = pd.to_datetime(df['crash_date'], format='%Y%m%d')
df['month'] = df['crash_date'].dt.month

# Создаем функцию для определения времени года
def get_season(month):
    if month in [12, 1, 2]:
        return 'Зима'
    elif month in [3, 4, 5]:
        return 'Весна'
    elif month in [6, 7, 8]:
        return 'Лето'
    else:
        return 'Осень'

df['season'] = df['month'].apply(get_season)

# Группируем данные по временам года
seasonal_stats = df.groupby('season').agg({
    'fatalities_amount': 'mean',
    'victims_amount': 'mean',
    'crash_date': 'count'  # количество ДТП
}).rename(columns={'crash_date': 'accident_count'}).reset_index()

# Упорядочиваем времена года
season_order = ['Зима', 'Весна', 'Лето', 'Осень']
seasonal_stats['season'] = pd.Categorical(seasonal_stats['season'], categories=season_order, ordered=True)
seasonal_stats = seasonal_stats.sort_values('season')

# Создаем фигуру с несколькими графиками
plt.figure(figsize=(14, 10))

# 1. График среднего количества ДТП по временам года
plt.subplot(2, 2, 1)
sns.barplot(x='season', y='accident_count', data=seasonal_stats, palette='coolwarm')
plt.title('Среднее количество ДТП по временам года')
plt.xlabel('Время года')
plt.ylabel('Количество ДТП')
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))  # только целые числа на оси Y

# 2. График средней тяжести ДТП
plt.subplot(2, 2, 2)
sns.barplot(x='season', y='fatalities_amount', data=seasonal_stats, palette='coolwarm')
plt.title('Среднее количество погибших по временам года')
plt.xlabel('Время года')
plt.ylabel('Количество погибших')

# 3. График среднего количества пострадавших
plt.subplot(2, 2, 3)
sns.barplot(x='season', y='victims_amount', data=seasonal_stats, palette='coolwarm')
plt.title('Среднее количество пострадавших по временам года')
plt.xlabel('Время года')
plt.ylabel('Количество пострадавших')

# 4. Круговая диаграмма распределения ДТП
plt.subplot(2, 2, 4)
plt.pie(seasonal_stats['accident_count'], 
        labels=seasonal_stats['season'],
        autopct='%1.1f%%',
        colors=sns.color_palette('coolwarm'),
        startangle=90,
        explode=(0.05, 0.05, 0.05, 0.05))
plt.title('Распределение ДТП по временам года')

plt.tight_layout()
plt.show()

# Дополнительный график: помесячная статистика
monthly_stats = df.groupby('month').agg({
    'fatalities_amount': 'mean',
    'victims_amount': 'mean',
    'crash_date': 'count'
}).rename(columns={'crash_date': 'accident_count'}).reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='accident_count', data=monthly_stats, 
             marker='o', color='b', label='Количество ДТП')
plt.title('Помесячная динамика количества ДТП')
plt.xlabel('Месяц')
plt.ylabel('Количество ДТП')
plt.xticks(range(1, 13), ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 
                         'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек'])
plt.grid(True)
plt.show()
```
